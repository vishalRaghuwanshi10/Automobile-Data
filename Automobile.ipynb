```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("Automobile_data.csv")
pd.set_option('display.max_rows', None)
del df1["index"]
```

```python
df1.head(5)
```

```python
df1.info() 
```

```python
df1["price"].fillna(0,inplace=True) # here we filled the null values with 0
```

```python
df1['price'] = df1['price'].astype("int64") # here we converted float into integer
df1["wheel-base"] = df1["wheel-base"].astype("int64")
df1["length"] = df1['length'].astype("int64")
```

```python
df1.info()
```

```python
# Here is the average price and average mileage of convertible cars:-


convertible_df = df1[df1['body-style'] == 'convertible']
average_price = convertible_df['price'].mean()
average_mileage = convertible_df['average-mileage'].mean()
average_df = pd.DataFrame({'Average Price': [average_price], 'Average Mileage': [average_mileage]})

average_df
```

```python
# in this cell i sorted cars by company and its price:-

sorted_cars = df1.sort_values(by=['company', 'price'], ascending=[True,False])
sorted_cars.reset_index(drop=True).head(5)
```

```python
# Data for toyota cars seprately

toyota_cars = df1[df1['company'] == 'toyota']
toyota_cars.sort_values(by=['price'],ascending=False).reset_index(drop=True)
```

```python
# company producing most expensive car:-

most_expensive_company = df1.loc[df1['price'].idxmax(), 'company']
max_price = df1['price'].max()
most_expensive_df = pd.DataFrame({'Company': [most_expensive_company], 'Price': [max_price]})

most_expensive_df
```

```python
# cars on basis of maximum length highest wheelbase and maximum horsepower:-

max_length_car = df1.loc[df1['length'].idxmax()]

max_wheelbase_car = df1.loc[df1['wheel-base'].idxmax()]

max_horsepower_car = df1.loc[df1['horsepower'].idxmax()]

e_cars_df = pd.DataFrame([max_length_car, max_wheelbase_car, max_horsepower_car])

e_cars_df = e_cars_df.drop_duplicates().reset_index(drop=True)
e_cars_df
```

```python
# company wise average price of cars in decending order:-

companywise_avg_price = df1.groupby('company')['price'].mean().sort_values(ascending=False)
companywise_avg_price = pd.DataFrame(companywise_avg_price,columns=['price'])


companywise_avg_price 
```

```python
#created a column name:"decision",If wheel-base is less than 100 and length is less than 175 then"don't buy" else "buy"

df1['decision'] = df1.apply(lambda x: 'don\'t buy' if (x['wheel-base'] < 100 and x['length'] < 175) else 'buy', axis=1)
df1.head(5)   


```

```python
# Selected cars whose decision is "buy":-

buy_cars = df1[df1['decision'] == 'buy']
buy_cars.reset_index(drop=True).head(5)  
```

```python
avg_price_buy_cars = buy_cars.groupby('company')['price'].mean()
avg_price_buy_cars = pd.DataFrame(avg_price_buy_cars)


avg_price_buy_cars
```

```python
writer = pd.ExcelWriter("Automobile.xlsx")

average_df.to_excel(writer,sheet_name = "Average_of_convertible_cars")
sorted_cars.to_excel(writer,sheet_name = "sorted_cars")
toyota_cars.to_excel(writer,sheet_name = "toyota_cars")
most_expensive_df.to_excel(writer,sheet_name = "most_expensive_car_company")
e_cars_df.to_excel(writer,sheet_name = "max")
companywise_avg_price.to_excel(writer,sheet_name = "companywise_avg_price")
df1.to_excel(writer,sheet_name = "don't buy  or buy")
buy_cars.to_excel(writer,sheet_name = "buy cars")
avg_price_buy_cars.to_excel(writer,sheet_name = "avg_price_buy_cars")

writer.save()
```
